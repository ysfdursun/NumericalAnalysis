# 16-bit IEEE 754 Representation

```python
import numpy as np
```

```python
np.sctypes
```

```python
x = np.float16(13.375)
```

```python
print(x)
```

```python
x_16 = np.linspace(0, 1, 101, dtype=np.float16)
x_32 = np.linspace(0, 1, 101, dtype=np.float32)
x_64 = np.linspace(0, 1, 101, dtype=np.float64)
```

```python
print(f"  16-bit     32-bit     64-bit ")
print(f"========== ========== ==========")
for a, b, c in zip(x_16, x_32, x_64):
    print(f"{a:.8f} {b:.8f} {c:.8f}")
```

```python
for i in range(10):
    print(f"{np.float16(0.17):.{i}f}")
```
